```python
def tokenize(text):
    text = text.lower()
    cleaned = ""
    for ch in text:
        if ch.isalnum() or ch.isspace():
            cleaned += ch

    tokens = []
    for word in cleaned.split(" "):
        if word != "":
            if word.endswith('s'):
                word = word[:-1]
            tokens.append(word)

    return tokens

tokens = tokenize("Hello World! This is a TEST sentence, with 123 numbers.")
print(tokens)
```

```python
def build_vocab(texts, min_freq):
    token_lists = [tokenize(text) for text in texts]
    
    freq = {}
    for tokens in token_lists:
        for w in tokens:
            freq[w] = freq.get(w, 0) + 1

    vocab_words = []
    unk_count = 0

    for w, c in freq.items():
        if c >= min_freq:
            vocab_words.append(w)
        else:
            unk_count += c 

    vocab_words.sort()
    vocab_words = ["<unk>"] + vocab_words

    word_to_index = {}
    index_to_word = {}

    for i, w in enumerate(vocab_words):
        word_to_index[w] = i
        index_to_word[i] = w

    unigram_freq = {}

    unigram_freq[0] = unk_count

    for i in range(1, len(vocab_words)):
        word = vocab_words[i]
        unigram_freq[i] = freq[word]

    return word_to_index, index_to_word, unigram_freq


texts = [
    "the cat sat on the mat",
    "the dog sat on the log",
    "cats and dogs"
]
word_to_index, index_to_word, unigram_freqs = build_vocab(texts, min_freq=2)
print(unigram_freqs)
print(word_to_index)
print(index_to_word)
```

```python
def make_cbow_samples(tokens, window_size, strict_window_size=True):
    samples = []

    for i in range(len(tokens)):
        center = tokens[i]
        
        if i == 0 or i == len(tokens) - 1: continue
        
        # check window size
        if strict_window_size:
            act_window_size = min(i + window_size + 1, len(tokens)) - max(0, i - window_size)
            if act_window_size < 2 * window_size + 1:
                continue

        context = []
        for j in range(i - window_size, i + window_size + 1):
            # center
            if j == i:
                continue
            
            # context
            if 0 <= j < len(tokens):
                context.append(tokens[j])

        if len(context) > 0:
            samples.append((context, center))

    return samples

print(make_cbow_samples(tokenize("the cat sat on the mat"), window_size=2))
print(make_cbow_samples(tokenize("the cat sat on the mat"), window_size=2, strict_window_size=False))
```

```python
def make_skipgram_samples(tokens, window_size):
    samples = []
    for i in range(len(tokens)):
        center = tokens[i]

        for j in range(i - window_size, i + window_size + 1):
            if j == i:
                continue
            
            if 0 <= j < len(tokens):
                samples.append((center, tokens[j]))

    return samples

print(make_skipgram_samples(tokenize("the cat sat on the mat"), window_size=2))
```

```python
def collate_fn(batch, word_to_idx, mode="cbow"):
    unk = word_to_idx["<unk>"]

    if mode == "cbow":
        contexts = []
        targets = []

        for context_words, target_word in batch:
            # convert words to indices
            ctx_indices = [word_to_idx.get(w, unk) for w in context_words]
            tgt_index = word_to_idx.get(target_word, unk)

            contexts.append(ctx_indices)
            targets.append(tgt_index)

        return contexts, targets


    elif mode == "skipgram":
        centers = []
        contexts = []

        for center_word, context_word in batch:
            cen_index = word_to_idx.get(center_word, unk)
            centers.append(cen_index)
            ctx_index = word_to_idx.get(context_word, unk)
            contexts.append(ctx_index)

        return centers, contexts


    else:
        raise ValueError("wrong mode name")
```

### **Unigram (for negative sampling) & Utility functions**

```python
import numpy as np
import math
import random
from typing import List, Tuple, Iterable, Dict

class UnigramTable:
    def __init__(self, freq: Dict[int, int], exponent: float = 0.75):
        vocab = sorted(freq.keys())
        counts = np.array([freq[i] for i in vocab], dtype=np.float64)
        probs = counts ** exponent
        probs /= probs.sum()
        self.vocab = np.array(vocab, dtype=np.int32)
        self.probs = probs

    def sample(self, k: int) -> np.ndarray:
        return np.random.choice(self.vocab, size=k, replace=True, p=self.probs)

# utility functions
def sigmoid(x):
    return 1.0 / (1.0 + np.exp(-x))

def safe_log(x, eps=1e-12):
    return np.log(x + eps)
```

### **CBOW Implementation**

```python
class CBOW:
    def __init__(self, vocab_size: int, emb_dim: int = 100, seed: int = 1234):
        self.vocab_size = vocab_size
        self.emb_dim = emb_dim

        # initialize embeddings
        rnd = np.random.RandomState(seed)
        bound = 0.5 / emb_dim
        self.input_emb = rnd.uniform(-bound, bound, (vocab_size, emb_dim)).astype(np.float64)
        self.output_emb = rnd.uniform(-bound, bound, (vocab_size, emb_dim)).astype(np.float64)

    def forward_softmax_loss(self, contexts_batch: List[List[int]], targets_batch: List[int]):
        batch_size = len(targets_batch)
        D = self.emb_dim
        V = self.vocab_size

        # Compute average of context embeddings
        context_vecs = np.zeros((batch_size, D), dtype=np.float64)
        for i, ctx in enumerate(contexts_batch):
            if len(ctx) == 0:
                continue
            context_vecs[i] = np.mean(self.input_emb[ctx], axis=0)

        # dot products
        logits = context_vecs.dot(self.output_emb.T)
        
        # softmax
        logits = logits - np.max(logits, axis=1, keepdims=True)
        exp_logits = np.exp(logits)
        probs = exp_logits / np.sum(exp_logits, axis=1, keepdims=True)  # (B,V)

        # loss
        # get the probability of the target word for each batch (2D indexing)
        true_probs = probs[np.arange(batch_size), targets_batch]
        loss = -np.sum(np.log(true_probs + 1e-12)) / batch_size

        # gradients
        probs[np.arange(batch_size), targets_batch] -= 1.0
        # gradient w.r.t output embeddings
        d_output = probs.T.dot(context_vecs) / batch_size
        # gradient w.r.t context vectors
        d_context = probs.dot(self.output_emb) / batch_size

        # gradient w.r.t to input embeddings
        # as context vector is the mean, its gradient is distributed evenly among input embeddings
        d_input = np.zeros_like(self.input_emb)
        for i, ctx in enumerate(contexts_batch):
            if len(ctx) == 0:
                continue
            # averaging
            grad_per_token = d_context[i] / len(ctx)
            for token in ctx:
                d_input[token] += grad_per_token

        return loss, d_input, d_output

    def forward_negative_sampling_loss(self, contexts_batch: List[List[int]], targets_batch: List[int],
                                       unigram_table: UnigramTable, neg_k: int = 5):
        batch_size = len(targets_batch)
        D = self.emb_dim
        V = self.vocab_size

        # compute context vectors
        context_vecs = np.zeros((batch_size, D), dtype=np.float64)
        ctx_lens = []
        for i, ctx in enumerate(contexts_batch):
            if len(ctx) == 0:
                ctx_lens.append(0)
                continue
            ctx_lens.append(len(ctx))
            context_vecs[i] = np.mean(self.input_emb[ctx], axis=0)

        # positive scores
        pos_out = self.output_emb[targets_batch]
        pos_scores = np.sum(pos_out * context_vecs, axis=1)

        # sample negative samples, neg_k negative samples for each positive sample
        # the shape will become (B, neg_k) for the entire batch
        neg_samples = unigram_table.sample(batch_size * neg_k).reshape(batch_size, neg_k)
        neg_out = self.output_emb[neg_samples]

        # compute losses
        pos_loss = -np.log(sigmoid(pos_scores) + 1e-12)
        # the following line is equivalent to using two for loops for the entire batch
        # the first argument notes the shapes of the operands and shape of the output of the production
        neg_scores = np.einsum('bkd,bd->bk', neg_out, context_vecs)
        neg_loss = -np.sum(np.log(sigmoid(-neg_scores) + 1e-12), axis=1)

        loss = np.mean(pos_loss + neg_loss)

        # gradients
        d_output = np.zeros_like(self.output_emb)
        d_input = np.zeros_like(self.input_emb)
        # positive gradient
        pos_grad_coeff = (sigmoid(pos_scores) - 1.0) / batch_size
        # update output embedding gradient for targets
        for i, tgt in enumerate(targets_batch):
            d_output[tgt] += pos_grad_coeff[i] * context_vecs[i]

        # negative gradients
        neg_sig = sigmoid(neg_scores)
        for i in range(batch_size):
            for k in range(neg_k):
                neg_tok = neg_samples[i, k]
                coeff = neg_sig[i, k] / batch_size
                d_output[neg_tok] += coeff * context_vecs[i]

        # gradient w.r.t context vector
        for i in range(batch_size):
            if ctx_lens[i] == 0:
                continue
            grad_vc = pos_grad_coeff[i] * self.output_emb[targets_batch[i]]  
            grad_vc += np.sum((neg_sig[i, :, None]) * neg_out[i], axis=0) / batch_size
            # distribute to input embeddings (averaged)
            per_token = grad_vc / ctx_lens[i]
            for tok in contexts_batch[i]:
                d_input[tok] += per_token

        return loss, d_input, d_output

    def step(self, d_input, d_output, lr=0.05):
        # SGD
        self.input_emb -= lr * d_input
        self.output_emb -= lr * d_output

    def most_similar(self, word_idx: int, topn=10):
        vec = self.input_emb[word_idx]
        
        # cosine similarity
        norms = np.linalg.norm(self.input_emb, axis=1) * (np.linalg.norm(vec) + 1e-12)
        sims = self.input_emb.dot(vec) / norms
        
        # exclude the given word
        sims[word_idx] = -np.inf
        top_idx = np.argsort(-sims)[:topn]
        
        return [top_idx, sims[top_idx]]
```

### **Skipgram Implementation**

```python
class SkipGram:
    def __init__(self, vocab_size: int, emb_dim: int = 100, seed: int = 1234):
        rnd = np.random.RandomState(seed)
        self.vocab_size = vocab_size
        self.emb_dim = emb_dim
        bound = 0.5 / emb_dim
        self.input_emb = rnd.uniform(-bound, bound, (vocab_size, emb_dim)).astype(np.float64)
        self.output_emb = rnd.uniform(-bound, bound, (vocab_size, emb_dim)).astype(np.float64)

    def forward_softmax_loss(self, centers: List[int], contexts: List[int]):
        B = len(centers)
        D = self.emb_dim
        V = self.vocab_size

        center_vecs = self.input_emb[centers]
        logits = center_vecs.dot(self.output_emb.T)
        
        # softmax
        logits = logits - np.max(logits, axis=1, keepdims=True)
        exp_logits = np.exp(logits)
        probs = exp_logits / np.sum(exp_logits, axis=1, keepdims=True)

        # loss
        true_probs = probs[np.arange(B), contexts]
        loss = -np.sum(np.log(true_probs + 1e-12)) / B

        # gradients
        probs[np.arange(B), contexts] -= 1.0
        # w.r.t output embeddings
        d_output = probs.T.dot(center_vecs) / B 
        # w.r.t center words
        d_center = probs.dot(self.output_emb) / B 
        # w.r.t input embeddings
        d_input = np.zeros_like(self.input_emb)
        for i, c_idx in enumerate(centers):
            d_input[c_idx] += d_center[i]

        return loss, d_input, d_output

    def forward_negative_sampling_loss(self, centers: List[int], contexts: List[int],
                                       unigram_table: UnigramTable, neg_k: int = 5):
        B = len(centers)
        D = self.emb_dim
        V = self.vocab_size

        center_vecs = self.input_emb[centers] 
        pos_out = self.output_emb[contexts] 
        pos_scores = np.sum(center_vecs * pos_out, axis=1)

        neg_samples = unigram_table.sample(B * neg_k).reshape(B, neg_k)
        neg_out = self.output_emb[neg_samples]

        pos_loss = -np.log(sigmoid(pos_scores) + 1e-12)
        # as explained and used in the CBOW implementation
        neg_scores = np.einsum('bd,bkd->bk', center_vecs, neg_out)
        neg_loss = -np.sum(np.log(sigmoid(-neg_scores) + 1e-12), axis=1)

        loss = np.mean(pos_loss + neg_loss)

        # gradients
        d_input = np.zeros_like(self.input_emb)
        d_output = np.zeros_like(self.output_emb)

        pos_coeff = (sigmoid(pos_scores) - 1.0) / B
        for i in range(B):
            tgt = contexts[i]
            d_output[tgt] += pos_coeff[i] * center_vecs[i]

        neg_sig = sigmoid(neg_scores)
        for i in range(B):
            for k in range(neg_k):
                tok = neg_samples[i, k]
                d_output[tok] += (neg_sig[i, k] / B) * center_vecs[i]

        for i in range(B):
            grad_vc = pos_coeff[i] * pos_out[i] + np.sum((neg_sig[i, :, None]) * neg_out[i], axis=0) / B
            d_input[centers[i]] += grad_vc

        return loss, d_input, d_output

    def step(self, d_input, d_output, lr=0.05):
        self.input_emb -= lr * d_input
        self.output_emb -= lr * d_output

    def most_similar(self, word_idx: int, topn=10):
        vec = self.input_emb[word_idx]
        norms = np.linalg.norm(self.input_emb, axis=1) * (np.linalg.norm(vec) + 1e-12)
        sims = self.input_emb.dot(vec) / norms
        sims[word_idx] = -np.inf
        top_idx = np.argsort(-sims)[:topn]
        return [top_idx, sims[top_idx]]
```

### **Training**

```python
import sys

def progress_bar(current, total, bar_length=40):
    fraction = current / total
    filled = int(fraction * bar_length)
    bar = "â–ˆ" * filled + " " * (bar_length - filled)
    percent = int(fraction * 100)
    sys.stdout.write(f"\r|{bar}| {percent:3d}% ({current}/{total})")
    sys.stdout.flush()
```

```python
import matplotlib.pyplot as plt

def train_cbow(
    model,
    train_loader,
    val_loader,
    word_to_idx,
    unigram_freq,
    epochs=5,
    lr=0.05,
    use_negative_sampling=True,
    neg_k=5,
):

    table = UnigramTable(unigram_freq) if use_negative_sampling else None
    train_losses = []
    val_losses   = []
    test_losses  = []
    for epoch in range(1, epochs+1):
        total_loss = 0.0
        n = 0
        num_batches = len(train_loader)

        for batch_id, (contexts_list, targets_list) in enumerate(train_loader, start=1):
            if use_negative_sampling:
                loss, d_in, d_out = model.forward_negative_sampling_loss(contexts_list, targets_list, table, neg_k)
            else:
                loss, d_in, d_out = model.forward_softmax_loss(contexts_list, targets_list)

            model.step(d_in, d_out, lr)
            total_loss += loss
            n += 1
            progress_bar(batch_id, num_batches)

        avg_train_loss = total_loss / max(1, n)
        train_losses.append(avg_train_loss)

        avg_val_loss = evaluate_cbow(
            model,
            val_loader,
            word_to_idx,
            unigram_freq,
            use_negative_sampling,
            neg_k
        )
        val_losses.append(avg_val_loss)

        print(f"[CBOW] Epoch {epoch}/{epochs}  "
              f"Train={avg_train_loss:.6f}  Val={avg_val_loss:.6f}")


    return train_losses, val_losses

def evaluate_cbow(
    model,
    eval_loader,
    word_to_idx,
    unigram_freq,
    use_negative_sampling=True,
    neg_k=5
):
    total = 0.0
    count = 0
    table = UnigramTable(unigram_freq) if use_negative_sampling else None
    for contexts_list, targets_list in eval_loader:
        if use_negative_sampling:
            loss, _, _ = model.forward_negative_sampling_loss(contexts_list, targets_list, table, neg_k)
        else:
            loss, _, _ = model.forward_softmax_loss(contexts_list, targets_list)

        total += loss
        count += 1

    return total / max(1, count)

def train_skipgram(
    model,
    train_loader,
    val_loader,
    word_to_idx,
    unigram_freq,
    epochs=5,
    lr=0.05,
    use_negative_sampling=True,
    neg_k=5,
):
    table = UnigramTable(unigram_freq) if use_negative_sampling else None

    train_losses = []
    val_losses = []
    test_losses = []
    for epoch in range(1, epochs+1):
        total_loss = 0.0
        n = 0
        num_batches = len(train_loader)
        
        for batch_id, (centers, contexts) in enumerate(train_loader, start=1):
            if use_negative_sampling:
                loss, d_in, d_out = model.forward_negative_sampling_loss(centers, contexts, table, neg_k)
            else:
                loss, d_in, d_out = model.forward_softmax_loss(centers, contexts)

            model.step(d_in, d_out, lr)
            total_loss += loss
            n += 1
            progress_bar(batch_id, num_batches)

        avg_train_loss = total_loss / max(1, n)
        train_losses.append(avg_train_loss)

        avg_val_loss = evaluate_skipgram(
            model,
            val_loader,
            word_to_idx,
            unigram_freq,
            use_negative_sampling,
            neg_k
        )
        val_losses.append(avg_val_loss)
        
        print(f"[SkipGram] Epoch {epoch}/{epochs}  "
              f"Train={avg_train_loss:.6f}  Val={avg_val_loss:.6f}")

    return train_losses, val_losses

def evaluate_skipgram(
    model,
    eval_loader,
    word_to_idx,
    unigram_freq,
    use_negative_sampling=True,
    neg_k=5
):
    total = 0.0
    count = 0
    table = UnigramTable(unigram_freq) if use_negative_sampling else None

    for centers, contexts in eval_loader:
        if use_negative_sampling:
            loss, _, _ = model.forward_negative_sampling_loss(centers, contexts, table, neg_k)
        else:
            loss, _, _ = model.forward_softmax_loss(centers, contexts)
        total += loss
        count += 1

    return total / max(1, count)

def plot_losses(train_losses, val_losses, test_loss, title="Loss"):
    plt.figure(figsize=(8,5))
    epochs = range(1, len(train_losses) + 1)
    plt.plot(epochs, train_losses, label="Train Loss")
    plt.plot(epochs, val_losses, label="Validation Loss")
    plt.axhline(y=test_loss, linestyle="--", label=f"Test Loss = {test_loss:.4f}")

    plt.xlabel("Epoch")
    plt.ylabel("Loss")
    plt.title(title)
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
import numpy as np

def embed_sentence(sentence, model):
    words = sentence.split()
    word_vectors = []

    for w in words:
        if w in model:
            word_vectors.append(model[w])
    
    if not word_vectors:  
        return np.zeros(model.vector_size)

    return np.mean(word_vectors, axis=0)
```
